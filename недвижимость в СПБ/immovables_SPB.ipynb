## Предобработка данных

### Пропуски

```python
data.isna().sum() # количество пропусков в солбцах
```

```python
data['is_apartment'] = data['is_apartment'].fillna(0) # заполнение пропущенных значений где это можно было сделать
data['balcony'] = data['balcony'].fillna(0)
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
data=data.dropna(subset=['floors_total'])
```

Причины наличия пропусков:
1. Человек который размещал оъявления не заполнял некоторые ячейки
2. При передаче файла данные могли пропасть
3. Если данные картографические то мог произойти баг

### Типы данных

```python
data.dtypes
```

```python
data['is_apartment']=data['is_apartment'].astype('bool')
data['first_day_exposition']=pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
name_columns = ['ponds_around3000', 'parks_around3000', 'balcony', 'floor','floors_total']
for name in name_columns:
    data[name] = data[name].astype('int64')
data.dtypes
```

1 ('ponds_around3000', 'parks_around3000', 'balcony,floor') не могут быть дробными значениями

##  Добавление в таблицу новых столбцов

```python
data.info()
```

```python
data['price_1_meter'] = data['last_price']/data['total_area']
```

```python
data['day_week'] = data['first_day_exposition'].dt.weekday # в столбце day_week дни недели от 0 до 6
data['month'] = data['first_day_exposition'].dt.month #в столбце month месяцы от 1 до 12
data['year'] = data['first_day_exposition'].dt.year #в столбце year года
```

```python
def floor_type(row):
    floor = row['floor']
    floors_total = row['floors_total']
    try:
        if floor == 1:
            return 'первый'
        if floor == 0:
            return 'цоколь'
        if floor == -1:
            return 'подвал'
        elif floor == floors_total:
            return 'последний'
        else:
            return 'другое'
    except:
        pass
data['floor_type'] = data.apply(floor_type, axis=1) # в столбце floor_type тип этажа(первый, последний или другое)
```

```python
data['cityCenters_nearest_km'] = round(data['cityCenters_nearest']/1000,0) #в столбце cityCenters_nearest_km расстояние до центра в км
```

### Средняя цена квадратного метра в 10-ти населенных пунктах

```python
state=data.pivot_table(index='locality_name', values='price_1_meter', aggfunc=['mean','count'])
state.columns=['mean_price','count']
```

```python
# график средней цены за квадратный метр в различных населенных пунктах
state.sort_values("count", ascending=False).head(10).sort_values(
    "mean_price", ascending=False
)["mean_price"].plot(
    kind="barh", 
    figsize=(15, 5),  
    grid=True,
    legend=True,
)                                                                  



```

По графику можно отметить что самые дешевые квартиры в Выборге а самые дорогие в городах Пушкин и СПБ
